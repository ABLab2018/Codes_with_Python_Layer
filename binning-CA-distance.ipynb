```python
import subprocess
import os
import glob

# Step 1: Save the Fortran program into a file
code = r"""
	parameter(np=4931,nres=56,nf=10,nbin=200)
	
	character filename(nf)*3,binfilename(nf)*11
	character bindexfile(nf)*13,bincountfilename(nf)*13
	real distance(nf,np,nres),sum1
	integer bin_index(nf,np,nres)
	
	bin_count=0
	bin_size=0.2
	
	open(1,file="nn-filename")
	open(3,file="binning-filename")
	open(8,file="bin-index-filename")
	open(9,file="bin-number-2LHC")
	open(13,file="bin-count-filename")
	open(14,file="bin-index-upto-nn10-2LHC")
	
	do i=1,nf
	read(1,"(a3)") filename(i)
	read(3,"(a11)") binfilename(i)
	read(8,"(a13)") bindexfile(i)
	read(13,"(a13)") bincountfilename(i)
	enddo
	
c	Read and distribute data into bins	
	
	do i=1,nf
	
	open(2,file="nn-calpha-dist-cal-2LHC/"//filename(i))
	
	do kk=1,np
	do j=1,nres
	read(2,*) distance(i,kk,j)
	enddo
	enddo
	
	enddo
	
	do kk=1,np
	do i=1,nf	
	do j=1,nres
	
	bin_index(i,kk,j) = int(distance(i,kk,j)/bin_size) + 1
   	   
	write(14,*) bin_index(i,kk,j)
	enddo
	enddo
	enddo
	
c	 Display bin counts
	
c	open(12,file="binning-file/"//bincountfilename(i))
	
c	do k=1,nbin
c	write(12,*) bin_count(i,k)
c	enddo
	
c	write(*,*) MAXVAL(distance)

	
	do k=1,nbin
	write(9,"(I3)") k
	enddo
	
	stop
	end

"""
with open("binning-CA-distance.f", "w") as f:
    f.write(code)
# Step 2: Compile the Fortran code using gfortran
subprocess.run(["gfortran", "binning-CA-distance.f", "-o", "binning-CA-distance"], check=True)

# Step 3: Run the Fortran executable
subprocess.run(["./binning-CA-distance"], check=True)
print("✅ Fortran program executed")
```
